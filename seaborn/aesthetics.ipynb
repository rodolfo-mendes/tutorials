# Controlling figure aesthetics

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

np.random.seed(sum(map(ord, 'aesthetics')))
````

````python
def sinplot(flip=1):
    x = np.linspace(0, 14, 1000)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
````

````python
sinplot()
````

````python
sns.set() # switch to seaborn default styles
sinplot()
````

````python
sns.axes_style() # returns a dictionary with style property values
````

````python
sns.plotting_context() # parameters to scale plot elements
````

````python
sns.set_style({'axes.grid': False})
sinplot()
````

````python
sns.set_style({'axes.grid': True})
sinplot()
````

## Seaborn figure styles

````python
sns.set_style('whitegrid')
sinplot()
````

The function ```sns.set_style()``` accepts two kind of arguments: one of the supported style names or a dictionary containing the style properties and their respective values.

````python
sns.set_style('whitegrid')
data = np.random.normal(size=(20,6)) + np.arange(6) / 2
sns.boxplot(data=data)
````

````python
sns.set_style('dark') # removes the grid from the darkgrid theme
sinplot()
````

````python
sns.set_style('white') # removes the grid from the whitegrid theme
sinplot()
````

````python
sns.set_style('ticks') # adds ticks to the axes
sinplot()
````

## Removing axes spines

````python
# axes spines may be removed simply calling the despine() function
sns.set_style('white')
sinplot()
sns.despine()
````

````python
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True) # the trim parameter limit the remaining spines to the range with data
````

````python
sns.set_style('whitegrid')
sns.boxplot(data=data, palette='deep')
sns.despine(left=True)
````

## Temporarily setting figure style

Its possible for seaborn set a figure style only for the current plot, so that, it is even possible to draw two chart on the same plot with different styles;

````python
plt.figure(figsize=(20, 5))

with sns.axes_style('darkgrid'):
    plt.subplot(1, 2, 1)
    sinplot()

plt.subplot(1, 2, 2)
sinplot(-1)
````

## Scalling plot elements

The scale of plot elements are controlled by a separate set of parameters.

````python
# reseting the parameters
sns.set()
````

There 4 preset contexts, in order of their relative size they are:
* paper
* notebook
* talk
* poster

The **notebook** preset is the default one.

````python
sns.set_context('paper')
sinplot()
````

````python
sns.set_context('talk')
sinplot()
````

````python
sns.set_context('poster')
sinplot()
````

````python
sns.set_context('notebook', font_scale=1.5, rc={'lines.linewidth': 2.5})
sinplot()
````
